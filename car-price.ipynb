```python
df5 = df4[df4["price"]<=500000]
df5
```

```python
df6 =  df5.drop(['hp'], axis='columns')
df6.head()
```

```python
df6.columns.tolist()
```

```python
dummies = pd.get_dummies(df6, prefix=['make','model','fuel','gear','offerType'],drop_first=True)
dummies
```

```python
dummies.columns.tolist()
```

```python
dummies.head()
```

```python
X = dummies.drop(['price'],axis="columns")
X.head()
```

```python
Y = dummies.price
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X,Y,cv=cv)
```

```python
X.columns.tolist()
```

```python
def get_the_index(df,col_name,col_value):
    index = np.where(df.columns == col_name + '_' + col_value)
    if index[0].size == 0:
        return -1
    else:
        return index[0][0]


def predict_car_price(mileage,year,make,model,fuel, gear,offerType):
    make_index = get_the_index(X, 'make',make)
    model_index = get_the_index(X, 'model',model)
    fuel_index = get_the_index(X, 'fuel',fuel)
    gear_index = get_the_index(X, 'gear',gear)
    offerType_index = get_the_index(X, 'offerType',offerType)
    x = np.zeros(len(X.columns))
    x[0] = mileage
    x[1] = year
    if make_index >=0:
        x[make_index] = 1
    if model_index >=0:
        x[model_index] = 1
    if fuel_index >= 0:
        x[fuel_index] = 1        
    if gear_index >=0:
        x[gear_index] = 1
    if offerType_index >=0:
        x[offerType_index] = 1    
    return x
```

```python
xv = predict_car_price(50000, 2020,'BMW','M5','Gasoline','Manual','Used')
```

```python
model.predict([xv])[0]
```
